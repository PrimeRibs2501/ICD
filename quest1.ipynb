```python
data_iris_without = pd.read_csv('/home/pedro/Documentos/024.2/icdados/data/data_base/iris.csv')
```

```python
data_iris_without
```

```python
df = pd.read_csv('data/iris.csv')

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
features = df.columns[:-1]
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=df)
    plt.title(f'Boxplot de {feature}')
    plt.xlabel('Espécies')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))
for i, col in enumerate(data_iris_without.columns[:4]):
    sns.boxplot(x='species', y=col, data=data_iris_without, ax=axes[i//2,i % 2]) 
    axes[i // 2, i % 2].set_title(f'Boxplot de {col}')
plt.show()
```
